```python
def choose_xo():
    player1 = ''
    player2 = ''
    player1 = input("Player 1: Which do you want to be, X or O? ").lower()
    if player1 == 'x':
        print("Player 1 will go first!")
        player2 = 'o'
    elif player1 == 'o':
        print("Player 2 will go first!")
        player2 = 'x'
    else:
        print("Please input 'X' or 'O'!!")
        player1, player2 = choose_xo()
    return [player1, player2]
```

```python
def display(status):
    print("   |   |   ")
    print(f" {status[2][0]} | {status[2][1]} | {status[2][2]}")
    print("---|---|---")
    print(f" {status[1][0]} | {status[1][1]} | {status[1][2]}")   
    print("---|---|---")
    print(f" {status[0][0]} | {status[0][1]} | {status[0][2]}")
    print("   |   |   ")  
```

```python
def result(status):
    win = False
    who_win = ''
    
    # check row if there is a bingo
    for row in status:
        #if any row has 3 x or o, assigned True to variable win
        if row.count('x') == 3:
            win = True
            who_win = 'x'
        elif row.count('o') == 3:
            win = True
            who_win = 'o'
        else:
            pass
        
    #check column if there is a bingo
    for n in range(3):
        #if any column has 3 x or o, assigned True to variable win
        column = [status[0][n],status[1][n],status[2][n]]
        if column.count('x') == 3:
            win = True
            who_win = 'x'
        elif column.count('o') == 3:
            win = True
            who_win = 'o'
        else:
            pass
    
    #check crossline (left-bottom to right-top) if there is a bingo
    crossln_lb2rt = [status[0][0], status[1][1], status[2][2]]
    if crossln_lb2rt.count('x') == 3:
        win = True
        who_win = 'x'
    elif crossln_lb2rt.count('o') == 3:
        win = True
        who_win = 'o'
    else:
        pass

    #check crossline (left-top to right-bottom) if there is a bingo
    crossln_lt2rb = [status[2][0], status[1][1], status[0][2]]
    if crossln_lt2rb.count('x') == 3:
        win = True
        who_win = 'x'
    elif crossln_lt2rb.count('o') == 3:
        win = True
        who_win = 'o'
    else:
        pass
    
    return [win, who_win]
```

```python
#function use to input the xo to the board
def input_ox(num, xo, status):
    update_status = status
    val_exist = False
    if num == 1:
        update_status[0][0] = xo
    elif num == 2:
        update_status[0][1] = xo
    elif num == 3:
        update_status[0][2] = xo
    elif num == 4:
        update_status[1][0] = xo
    elif num == 5:
        update_status[1][1] = xo
    elif num == 6:
        update_status[1][2] = xo
    elif num == 7:
        update_status[2][0] = xo
    elif num == 8:
        update_status[2][1] = xo
    elif num == 9:
        update_status[2][2] = xo
    return update_status
```

```python
def replay():
    ans = input("Do you want to replay? (Y/N)").lower()
    if ans == 'y':
        return True
    elif ans == 'n':
        return False
    else:
        replay()
```

```python
def tictactoe():
    end_game = True
    #greeting
    print("Welcome to the tic tac toe game")
    
    while end_game:
        #player choose to be x or o
        player1, player2 = choose_xo()
        player = {player1:'Player1', player2:'Player2'}
        playgame = input("Ready for the Game? (Y/N)?").lower()
        if playgame == 'y': 
            
            #creat a board for tic tac toe
            status = [[' ',' ',' '],[' ',' ',' '],[' ',' ',' ']]
        
            #creat a winning status variable
            winning = [False, '']
            
            #game play, start from x, ask player to input value
            for round in range(9):
                if round%2 == 0:
                    display(status)
                    num = int(input(f'Turn: {player["x"]}\nChoose a position (1-9):'))
                    status = input_ox(num, 'x', status)
                    winning = result(status)
                    if winning[0]:
                        display(status)
                        break
                    
                else:
                    display(status)
                    num = int(input(f'Turn: {player["o"]}\nChoose a position (1-9):'))
                    status = input_ox(num, 'o', status)
                    winning = result(status)
                    if winning[0]:
                        display(status)
                        break
            
            #announce the result
            if winning[0]:
                winner = player[winning[1]]
                print(f'Winner is {winner}!!!')
            else:
                print("This is a tie game!!")
            
            end_game = replay()
                
        elif playgame == 'n':
            #close the game
            end_game = False
        
        else:
            print("Please input a valid value!!")
```
